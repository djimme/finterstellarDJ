```python
import pandas as pd
```

```python
period = pd.period_range(start='2022-01-13 00:00:00', end='2022-01-13 02:30:00', freq = '30T')
data = {'col1' :[1,2,3,4,5,6], 'col2':period}
idx = ['row1', 'row2', 'row3', 'row4', 'row5', 'row6']

df = pd.DataFrame(data = data, index=idx)

df
```

```python
df['col1'].rolling(window=3).sum()
```

```python
df
```

```python
import numpy as np
```

```python
df['left'] = np.nan
df['right'] = np.nan
df['both'] = np.nan
df['neither'] = np.nan
df
```

```python
df['left'] = df['col1'].rolling(window=3, closed='left').sum()
df['right'] = df['col1'].rolling(window=3, closed='right').sum()
df['both'] = df['col1'].rolling(window=3, closed='both').sum()
df['neither'] = df['col1'].rolling(window=3, closed='neither').sum()
df
```

```python
df['nn']= df['col1'].rolling(window=3).sum()
```

```python
df
```

```python
df['neither'] = df['col1'].rolling(window=3,  min_periods=2, closed='neither').sum()
```

```python
df
```

```python
df['left_c'] = np.nan
df['right_c'] = np.nan
df['both_c'] = np.nan
df['neither_c'] = np.nan
df
```

```python
df['left_c'] = df['col1'].rolling(window=3, closed='left', center=True).sum()
df['right_c'] = df['col1'].rolling(window=3, closed='right', center=True).sum()
df['both_c'] = df['col1'].rolling(window=3, closed='both', center=True).sum()
df['neither_c'] = df['col1'].rolling(window=3, closed='neither', min_periods=2, center=True).sum()
df
```
